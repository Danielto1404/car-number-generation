```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
path = "../../../dataset/data.csv"
data = pd.read_csv(path)
```

```python
data.columns
```

```python
names = data.car_name.astype(str).apply(lambda s: s.split()[0].upper())
```

```python
cloud = WordCloud(max_font_size=50, max_words=70, background_color="white")
cloud.generate_from_text(" ".join(names))
```

```python
plt.figure(figsize=(16, 12))
plt.axis("off")
plt.imshow(cloud, interpolation="bilinear")
```
